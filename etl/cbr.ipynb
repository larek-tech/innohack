# Dependencies

```python
import httpx
import pathlib
import time
import random
import os
import sys
import asyncio
```

# Config

```python

DATA_PATH= pathlib.Path("../data/files")
```

```python
law_acts_base_url = "https://www.cbr.ru/Crosscut/LawActs/File/{id}" # id - int < 9885 valid on 27.10.24
```

```python
stats = {
}

async def download_pdf(path: pathlib.Path, url: str) -> None:
    timeout = random.randint(1, 5) / 10
    await asyncio.sleep(timeout)

    r = httpx.get(url, timeout=20)
    if r.status_code in stats:
        stats[r.status_code].append(url)
    else:
        stats[r.status_code] = [url]
    
    print(r.status_code, url, timeout)
    if r.status_code != 200:
        return
    name = url.split("/")[-1]
    with open(path  / f"{name}.pdf", 'wb') as f:
        f.write(r.content)
```

```python
if not os.path.isdir(DATA_PATH / "law_acts"):
    os.mkdir(DATA_PATH / "law_acts")
```

```python
jobs = [
    download_pdf(DATA_PATH / "law_acts", law_acts_base_url.format(id=_id))
    for _id in range(157, 10000)
]
```

```python
await asyncio.gather(*jobs)
```

```python
code_stats = {
    k: len(v)
    for k, v in stats.items()
}
```
